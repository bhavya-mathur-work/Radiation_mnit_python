```python
import pandas as pd
import matplotlib.pyplot as plt

d = pd.read_csv('/content/Bhavya Mathur - dose_vs_time.csv')

print(d)
```

```python
d['timestamp'] = pd.to_datetime(d['timestamp'])
```

```python
plt.figure(figsize=(12, 6))
plt.bar(d['timestamp'], d['dose'], width=1.0, color='steelblue')
plt.xlabel('Time', fontsize=12)
plt.ylabel('Absorbed Dose Rate', fontsize=12)
plt.title('Absorbed Dose Rate vs Time', fontsize=14)
plt.tight_layout()
plt.show()
```
