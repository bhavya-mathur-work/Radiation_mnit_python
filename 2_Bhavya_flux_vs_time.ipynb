```python
import pandas as pd
import matplotlib.pyplot as plt

f = pd.read_csv('/content/Bhavya Mathur - flux_vs_time.csv')

print(f)
```

```python
f['timestamp'] = pd.to_datetime(f['timestamp'])
```

```python
plt.figure(figsize=(10, 6))
plt.bar(f['timestamp'], f['flux'], color='green', linewidth=1.5)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Flux', fontsize=12)
plt.title('Flux vs Time', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
