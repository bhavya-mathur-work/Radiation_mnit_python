```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/Bhavya Mathur - doseEq.csv')  # Adjust the path as needed

df['timestamp'] = pd.to_datetime(df['timestamp'])


plt.figure(figsize=(12, 6))
plt.plot(df['timestamp'], df['dose_equivalent_rate'], color='blue', linewidth=1)

plt.xlabel('Time', fontsize=12)
plt.ylabel('Dose Equivalent Rate (µSv/h)', fontsize=12)
plt.title('Dose Equivalent Rate vs Time (ALTEA - ISS)', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```
