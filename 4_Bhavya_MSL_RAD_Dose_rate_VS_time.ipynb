```python
import pandas as pd

import matplotlib.pyplot as plt

url = ("https://visualization.osdr.nasa.gov/radlab/api/?spacecraft=&module=&instrument=&timestamp=&format=csv&timestamp%3E=2007-07-03T10:26&timestamp%3C=2024-06-17T11:59&instrument_id=MSL-RAD-Surface&absorbed_dose_rate=&")

d = pd.read_csv(url)

print(d)

```

```python
d['timestamp'] = pd.to_datetime(d['timestamp'])

plt.plot(d['timestamp'] , d['absorbed_dose_rate'])

plt.xlabel('Time', fontsize=12)
plt.ylabel('Absorbed Dose Rate', fontsize=12)
plt.title('Absorbed Dose Rate from MSL-RAD on Curiosity(Linear)', fontsize=14)
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(d['timestamp'], d['absorbed_dose_rate'], color='purple', linewidth=0.6)
plt.yscale('log')  # Set log scale for Y-axis

plt.xlabel('Time', fontsize=12)
plt.ylabel('Absorbed Dose Rate (log scale)', fontsize=12)
plt.title('Absorbed Dose Rate (Log Scale) from MSL-RAD on Curiosity', fontsize=14)
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
