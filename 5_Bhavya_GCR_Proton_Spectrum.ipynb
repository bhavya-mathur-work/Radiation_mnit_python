```python
import pandas as pd
import matplotlib.pyplot as plt


d = pd.read_csv('/content/Bhavya Mathur - gcr_flux.csv')


```

```python
print(d)
```

```python
d['Energy'] = pd.to_numeric(d['Energy'], errors='coerce')
d['proton'] = pd.to_numeric(d['proton'], errors='coerce')


d.dropna(inplace=True)

plt.figure(figsize=(10, 6))
plt.loglog(d['Energy'], d['proton'], marker='o', linestyle='-', color='green')

plt.title("GCR Spectrum: Proton Flux vs Energy", fontsize=14)
plt.xlabel("Energy (MeV)", fontsize=12)
plt.ylabel("Proton Flux (arbitrary units)", fontsize=12)
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
