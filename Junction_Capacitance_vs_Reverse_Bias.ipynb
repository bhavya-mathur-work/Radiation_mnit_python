```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
q = 1.6e-19  # Coulombs
epsilon_0 = 8.85e-14  # F/cm
epsilon_r = 11.7
epsilon_s = epsilon_r * epsilon_0  # F/cm

N_D = 1e15  # cm^-3, donor concentration

# Voltage range from 0.1 to 100 V. Avoid 0 to prevent division by zero
V = np.linspace(0.1, 100, 500)

# Calculate depletion width W (in cm)
W = np.sqrt((2 * epsilon_s * V) / (q * N_D))

# Calculate junction capacitance per unit area C_j (F/cm^2)
C_j = epsilon_s / W

# Convert capacitance to pF/cm^2 (1 F = 1e12 pF)
C_j_pF = C_j * 1e12

# Plotting
plt.figure(figsize=(8,5))
plt.plot(V, C_j_pF, label='C$_J$ (pF/cm$^2$)', color='blue')
plt.xlabel('Reverse Bias Voltage V (Volts)')
plt.ylabel('Junction Capacitance $C_J$ (pF/cm$^2$)')
plt.title('Junction Capacitance vs. Reverse Bias Voltage for n-type Silicon Detector')
plt.legend()
plt.grid(True)
plt.show()
```
